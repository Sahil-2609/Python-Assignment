## 2. Describe the role of predefined keywords in Python and provide examples of how they are used in a program.

##  Predefined Keywords in Python

Python uses reserved keywords that cannot be used as identifiers. These keywords have special meanings and are integral to the language's syntax. Examples include `if`, `else`, `def`, and `class`.

```python
# Example of using keywords in Python
def check_number(x):
    if x > 0:
        return "Positive"
    elif x == 0:
        return "Zero"
    else:
        return "Negative"

print(check_number(10))
```

## 3. Compare and contrast mutable and immutable objects in Python with examples.

##   Mutable vs Immutable Objects

**Mutable Objects**: These can be changed after their creation. Examples include lists and dictionaries.

**Immutable Objects**: These cannot be changed once created. Examples include tuples and strings.

```python
# Example of mutable object (list)
my_list = [1, 2, 3]
my_list.append(4)  # The list is modified
print("Mutable List:", my_list)

# Example of immutable object (tuple)
my_tuple = (1, 2, 3)
# Uncommenting the next line will raise an error
# my_tuple.append(4)
print("Immutable Tuple:", my_tuple)
```

## 4. Discuss the different types of operators in Python and provide examples of how they are used.

##  Types of Operators in Python

Python includes several types of operators:

- **Arithmetic Operators**: `+`, `-`, `*`, `/`, `%`, `**`, `//`
- **Comparison Operators**: `==`, `!=`, `>`, `<`, `>=`, `<=`
- **Logical Operators**: `and`, `or`, `not`
- **Assignment Operators**: `=`, `+=`, `-=`, `*=`, `/=`

```python
# Example of different types of operators
a = 10
b = 5

# Arithmetic Operators
print("Addition:", a + b)
print("Exponentiation:", a ** b)

# Comparison Operators
print("Greater than:", a > b)

# Logical Operators
print("Logical AND:", a > 5 and b < 10)
```

## 5. Explain the concept of type casting in Python with examples.

##  Type Casting in Python

Type casting allows converting a variable from one type to another. Common conversions include `int()`, `float()`, `str()`, and `list()`.

```python
# Example of type casting
x = 10  # Integer
y = float(x)  # Convert to float
z = str(y)  # Convert to string

print("Float:", y)
print("String:", z)
```

## 6. How do conditional statements work in PYthon? Illustrate with examples.

##  Conditional Statements in Python

Conditional statements are used to execute code based on certain conditions. Python uses `if`, `elif`, and `else` for conditional logic.

```python
# Example of conditional statements
age = 20

if age < 18:
    print("Minor")
elif age < 65:
    print("Adult")
else:
    print("Senior")
```

## 7. Describe the different types of loops in Python and their uses cases with example.

##  Types of Loops in Python

Python supports two primary types of loops:

- **`for` Loop**: Iterates over a sequence (e.g., list, tuple, string).
- **`while` Loop**: Repeats as long as a condition is true.

```python
# Example of a for loop
print("For Loop:")
for i in range(5):
    print(i)

# Example of a while loop
print("While Loop:")
count = 0
while count < 5:
    print(count)
    count += 1
```
